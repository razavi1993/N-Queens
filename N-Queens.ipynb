```julia
global N = 8
board = fill(".", N, N)
```

```julia
function isSafe(board, row, col)
    for x in range(1, col-1)
        if board[row,x] == "ðŸ‘‘"
            return false
        end
    end
    for (x, y) in zip(reverse(range(1, row-1)), reverse(range(1, col-1)))
        if board[x,y] == "ðŸ‘‘"
            return false
        end
    end
    for (x, y) in zip(range(row+1, N), reverse(range(1, col-1)))
        if board[x,y] == "ðŸ‘‘"
            return false
        end
    end
    return true
end
```

```julia
function solveNQueens(board, col, list)
    if col == N+1
        list = push!(list, copy(board))
        return 
    end
    for i in range(1, N)
        if isSafe(board, i, col)
            board[i,col] = "ðŸ‘‘"
            solveNQueens(board, col+1, list)
            board[i,col] = "."
        end
    end
end
```
